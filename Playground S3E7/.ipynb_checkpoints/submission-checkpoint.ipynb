```python
df = train.drop_duplicates(subset=train.columns.tolist(), keep=False)
```

```python
df.head()
```

```python
df.shape
```

```python
from pandas.tseries.holiday import USFederalHolidayCalendar

def fe(df):
        # Fix date anomalies (pd.to_datetime throws parsing error for some days, see anomalies section).
        df['arrival_year_month'] = pd.to_datetime(df['arrival_year'].astype(str) + df['arrival_month'].astype(str), format='%Y%m')
        df.loc[df.arrival_date > df.arrival_year_month.dt.days_in_month, 'arrival_date'] = df.loc[df.arrival_date > df.arrival_year_month.dt.days_in_month, 'arrival_year_month'].dt.days_in_month
        df.drop(columns='arrival_year_month', inplace=True)
        
        return df
```

```python
df = fe(df)
```

```python
df.sample(10)
```

# Preparing the model

#### importing modules required for model

```python
from sklearn.metrics import auc, accuracy_score, confusion_matrix, mean_squared_error
from sklearn.model_selection import cross_val_score, GridSearchCV, KFold, RandomizedSearchCV, train_test_split

import xgboost as xgb
```

```python
## X and Y data set
y = df.booking_status
X = df.drop(['id','booking_status'], axis=1) 
```

```python
## train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

# XGBoost

## Hyper parameters

```python
params = {
    "learning_rate" : [0.05, 0.10, 0.15, 0.20, 0.25, 0.30],
    "max_depth" : [3, 4, 5 ,6, 7, 8, 10, 12, 15],
    "min_child_weight" : [1, 3, 5, 7],
    "gamma" : [0.0, 0.1, 0.2, 0.3, 0.4],
    "colsample_bytree" : [0.3, 0.4, 0.5, 0.7]
}
```

```python
## creating xgboost model
xgb_model = xgb.XGBClassifier()
```

## Cross validation

```python
kfold_validation = KFold(10)
```

#### Tuning hyper parameters using Randomized Search CV

```python
 random_search = RandomizedSearchCV(xgb_model, param_distributions = params, n_iter = 5, scoring='roc_auc', n_jobs= -1, cv=kfold_validation, verbose=3)
```

```python
random_search.fit(X_train, y_train)
```

## Best params

```python
random_search.best_params_
```

```python
xgb_model = xgb.XGBClassifier(min_child_weight=3, max_depth=8, learning_rate=0.2,gamma=0.0, colsample_bytree=0.5)
```

```python
# xgb_model = xgb.XGBClassifier(objective="binary:logistic", random_state=42, eval_metric="auc")

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
xgb_model.fit(X_train, y_train, early_stopping_rounds=5, eval_set=[(X_test, y_test)])

y_pred = xgb_model.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
# result 
result = cross_val_score(xgb_model,X,y, cv=kfold_validation)
```

```python
## displaying the result 
print(result)

## mean
print(np.mean(result))
```

```python
print("best score: {0}, best iteration: {1}, best ntree limit {2}".format(xgb_model.best_score, xgb_model.best_iteration, xgb_model.best_ntree_limit))
```

## Testing the model

```python
test = fe(test)
```

```python
test_df=test.drop(["id"], axis=1)
```

```python
predictions = xgb_model.predict(test_df)
```

# Submission

```python
# Creating a dataframe in required format for submission
submission=pd.DataFrame()
submission["id"]=test["id"]
submission["booking_status"]=predictions
submission.head()
```
